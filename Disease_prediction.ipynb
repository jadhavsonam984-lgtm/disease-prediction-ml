```python
data=pd.read_csv('Disease_symptom_and_patient_profile_dataset.csv')
```

```python
# Count samples per disease
disease_counts = data['Disease'].value_counts()

# Select diseases having >= 25 samples
valid_diseases = disease_counts[disease_counts >= 10].index

# Filter dataset
data = data[data['Disease'].isin(valid_diseases)]
```

```python
x=data.drop('Disease',axis=1)
y=data['Disease']
```

```python
x
```

```python
print("Rows:", data.shape[0])
print("Features:", data.shape[1])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y = le.fit_transform(y)
```

```python
y
```

```python
print(le.classes_)
```

```python
x=pd.get_dummies(x).astype(int)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(
    x,y,
    test_size=0.1,
    stratify=y,
    random_state=42
)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)
```

```python
from sklearn.svm import SVC
svc_model = SVC(
    kernel='rbf',     # best general-purpose kernel
    C=1.0,            # regularization
    gamma='scale',
    random_state=42
)
svc_model.fit(x_train, y_train)
print("Train Accuracy:", svc_model.score(x_train, y_train))
print("Test Accuracy:", svc_model.score(x_test, y_test))
```

```python
model=gb.fit(x_train, y_train)
print("Train:", gb.score(x_train, y_train))
print("Test:", gb.score(x_test, y_test))
```

```python
y_pred=model.predict(x_test)
print(f"Predicted label for the first test sample: {le.inverse_transform([y_pred[0]])[0]}")
print(f"Actual label for the first test sample: {le.inverse_transform([y_test[0]])[0]}")
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print(acc)
```

```python
import numpy as np

new_data_sample = x_test.iloc[0].values.reshape(1, -1)
new_data = model.predict(new_data_sample)
print(f"Predicted label for the first test sample: {le.inverse_transform(new_data)}")
```
